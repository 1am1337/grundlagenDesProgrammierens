```python
x = np.arange(100) / 100

plt.plot(x * 2 * np.pi, np.sin(x * 2 * np.pi), '.-')
plt.xlabel('Radianten')
plt.ylabel('Amplitude')
plt.show()
```

```python
x = np.arange(100) / 100

plt.plot(x, np.sin(x * 2 * np.pi), '.-')
plt.xlabel('Zeit')
plt.ylabel('Amplitude')
plt.show()
```

```python
x = np.arange(44100) / 44100

freqs = np.arange(100) # in Hz

for freq in freqs:

    plt.title(f'Frequenz: {freq}')
    plt.plot(x, np.sin(freq * x * 2 * np.pi), '.-')
    plt.xlabel('Zeit')
    plt.ylabel('Amplitude')
    plt.show()
```
